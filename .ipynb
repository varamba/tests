В таблице books все ок, пропусков нет, дупликатов нет, единственное проверю тип данных.

В таблице возвратов есть дупликаты - удалю их.

Личный комментарий:
В некоторых колонках имеются странные значения - возможно ошибка при вводе данных или сбой при выгрузке - все это нормально, реальные данные выглядят похоже - исправлять их или удалять не стоит, тк это может повлиять на общую картину, если какие-то данные будут мешать при расчете - то тогда попытаюсь это исправить, пока что радотаю с явными дупликатами/пропусками/неверными типами данных, остальное требует более глубокого анализа.

```python
#проверю столбец copy_id тк у него должен быть тип int, а не object
display(books['copy_id'].unique())
```

Есть неправильные значения,такие как 0_7 или 199_1, которые могут повлиять на анализ, тк этот столбец имеет важное значение - заменю их на nan.

```python
#меняю типы данных
books['lib_date'] = pd.to_datetime(books['lib_date'], errors = 'coerce').dt.date
readers['born_date'] = pd.to_datetime(readers['born_date'], errors = 'coerce').dt.date
issues['return_date'] = pd.to_datetime(issues['return_date'], errors = 'coerce').dt.date
issues['issue_date'] = pd.to_datetime(issues['issue_date'], errors = 'coerce').dt.date

books['copy_id'] = pd.to_numeric(books['copy_id'], errors='coerce')
books['copy_id'] = books['copy_id'].astype('Int64')
```

```python
#удаляю дупликаты
issues = issues.drop_duplicates().reset_index(drop=True)
issues.duplicated().sum()
```

```python
#добавлю столбец "год издания"
delta = dt.timedelta(days=5*365)
books['pub_date'] = books['lib_date'] - delta
books['pub_date'] = pd.to_datetime(books['pub_date']).dt.date
books.head()
```

Предобработка завершена - можно выполнять запросы по заданиям.

```python
# install
!apt install postgresql postgresql-contrib &>log
!service postgresql start
!sudo -u postgres psql -c "CREATE USER root WITH SUPERUSER"
# set connection

%load_ext sql
%config SqlMagic.feedback=False 
%config SqlMagic.autopandas=True
%sql postgresql+psycopg2://@/postgres

from sqlalchemy import create_engine, text
engine = create_engine('postgresql+psycopg2://@/postgres')
conn = engine.connect()
```

```python
books.to_sql('books_q',conn,index=False,if_exists='replace',method='multi')
readers.to_sql('readers_q',conn,index=False,if_exists='replace',method='multi')
issues.to_sql('issues_q',conn,index=False,if_exists='replace',method='multi')
```

1. Найти города, в которых в 2016 году было издано больше всего книг:

```python
query_one = text("""
select city, count(book_id)as qty
from books_q
where extract(year from pub_date::date) = 2016
group by city

""")
```

```python
res_one = pd.read_sql_query(query_one,conn)
res_one
```

Получилось 34 города, которые издали всего по 1 книге за 2016 год, каких-то лидеров нет.

2. Вывести количество экземпляров книг «Война и мир» Л.Н.Толстого, которые находятся в библиотеке:

```python
query_two = text("""
select count(b.title)
from books_q as b
join issues as i using(copy_id)
join readers as r using(ticket_id)
where return_date is not null
and title = 'Война и мир'
""")
```

```python
res_two = pd.read_sql_query(query_two,conn)
res_two
```

В базе данных нет такой книги.

3. Найти читателей, которые за последний месяц брали больше всего книг в библиотеке. При выводе выполнить сортировку читателей по возрасту (от молодых к старшим):

```python
query_three = text("""
SELECT name, born_date ,count(i.ticket_id)
FROM readers r
join issues i using(ticket_id)
where extract(month from i.issue_date::date) = extract(month from current_date)
group by name, born_date
order by count(i.ticket_id) desc, r.born_date asc
""")
```

```python
res_three = pd.read_sql_query(query_three,conn)
res_three
```

Всего 41 читателей брали книги в последний месяц, из них всего 2 взяли две книги, остальные по одной.

Реализовать алгоритм для подсчета длины максимальной последовательности подряд идущих положительных чисел в массиве, состоящем из произвольного количества целых чисел (как положительных, так и отрицательных). 

Алгоритм должен получать на вход массив чисел и возвращать размер максимальной последовательности положительных чисел на выходе.


```python
num_in = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/КОРУС тест/Задание 2/numbers.csv')
num_out = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/КОРУС тест/Задание 2/numbers_out.csv')
```

```python
display(num_in)
display(num_out)
```

```python
nums = num_in['value'].tolist()
nums
```

```python
counter = 0
max_lenght = 0
for i in nums:
    if i > 0:
        counter += 1
        if counter > max_lenght:
            max_lenght = counter
    elif i <= 0:
        counter = 0        
print(max_lenght)
```

а. множество операций, состоящее из структур/кортежей вида <номер_операции, год, месяц, день, номер_подразделения, прибыль>.
